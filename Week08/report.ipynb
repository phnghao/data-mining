# 3. Báo cáo môn khai thác dữ liệu tuần 8

## 1. Tóm tắt phần code cài đặt thuật toán Mahalanobis K-Means
Thuật toán được xây dựng thông qua class `MahalanobisKMeans` hoạt động theo các bước chính sau:
* **Khởi tạo (Initialization):** Lựa chọn ngẫu nhiên $K$ điểm trong tập dữ liệu làm các tâm cụm ban đầu. Gán nhãn các điểm dữ liệu vào các tâm này dựa trên khoảng cách Euclidean cơ bản để tạo tiền đề. 
* **Tính toán ma trận hiệp phương sai:** Dựa trên các cụm khởi tạo, tính toán ma trận hiệp phương sai (covariance matrix) cho từng cụm. Nếu cụm chỉ có 1 điểm, sử dụng ma trận đơn vị.
* **Vòng lặp tối ưu (EM-like step):**
  * **Bước gán nhãn (Assignment):** Thay vì dùng khoảng cách Euclidean, tính toán bình phương khoảng cách Mahalanobis từ mỗi điểm dữ liệu đến các tâm cụm thông qua ma trận nghịch đảo của ma trận hiệp phương sai. Điểm dữ liệu được gán vào cụm có khoảng cách Mahalanobis nhỏ nhất.
  * **Bước cập nhật (Update):** Dựa vào nhãn mới, tính toán lại tọa độ vector trung bình (tâm cụm mới) và cập nhật lại ma trận hiệp phương sai cho từng cụm.
* **Điều kiện dừng:** Quá trình lặp lại liên tục cho đến khi nhãn của tất cả các điểm dữ liệu không còn sự thay đổi hoặc đạt đến số vòng lặp tối đa `max_iter`.

## 2. Nhận xét về thuật toán K-Means, GMM và Mahalanobis K-Means

**1. K-Means (Euclidean):**
* **Đặc điểm:** Hoạt động theo cơ chế gán cứng (hard assignment), sử dụng độ đo khoảng cách Euclidean truyền thống.
* **Nhược điểm:** Mặc định giả định rằng các cụm dữ liệu có dạng hình cầu tròn trịa và có phương sai tương đương nhau. Đường ranh giới phân chia giữa các cụm thường là các đường thẳng phân thủy vuông góc. Do đó, K-Means xử lý rất kém đối với các cụm dữ liệu có hình dáng kéo dài (elip), các cụm có mật độ khác nhau hoặc khi dữ liệu có nhiều nhiễu ở vùng biên.

**2. Gaussian Mixture Model (GMM):**
* **Đặc điểm:** Là mô hình dựa trên xác suất phân phối chuẩn, hoạt động theo cơ chế gán mềm (soft assignment) - tính toán xác suất một điểm thuộc về cụm nào đó. 
* **Ưu điểm:** Khai thác triệt để ma trận hiệp phương sai, giúp mô hình nhận diện cực kỳ linh hoạt hình dáng của các cụm dữ liệu. GMM hoạt động đặc biệt xuất sắc và gom cụm chính xác đối với các dữ liệu bị kéo dài, biến dạng thành hình elip hoặc các cụm có ranh giới chồng lấn đan xen nhau.

**3. Mahalanobis K-Means:**
* **Đặc điểm:** Là sự lai tạo mang ưu điểm của cả hai thuật toán trên. Vẫn duy trì cơ chế gán cứng (mỗi điểm chỉ thuộc về 1 cụm duy nhất) giúp dễ triển khai như K-Means, nhưng thay đổi độ đo thành khoảng cách Mahalanobis.
* **Ưu điểm:** Khoảng cách Mahalanobis có xét đến phương sai và độ tương quan đa chiều của các đặc trưng. Nhờ đó, ranh giới phân cụm không còn là đường thẳng vuông góc mà có thể uốn theo chiều phân bố của dữ liệu. Nó khắc phục được nhược điểm của K-Means truyền thống, đem lại chất lượng gom nhóm các cụm hình elip gần tương đương với GMM mà không cần sử dụng cơ chế tính toán kỳ vọng xác suất phức tạp. 
* **Lưu ý:** Cần xử lý ngoại lệ khi ma trận hiệp phương sai bị suy biến (không thể nghịch đảo), thuật toán sẽ tạm thời lui về sử dụng ma trận đơn vị (tương đương với khoảng cách Euclidean).
