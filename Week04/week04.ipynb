## 1. Giới thiệu
Trong bài thực hành này, thuật toán Apriori được sử dụng để khai phá luật kết hợp từ tập dữ liệu giao dịch. Bài toán được triển khai theo hai cách:
- Sử dụng thư viện có sẵn (mlxtend)
- Tự cài đặt thuật toán Apriori


## 2. Mô tả phương pháp

### 2.1. Tiền xử lý dữ liệu
Dữ liệu ban đầu chứa các giá trị NaN, do đó cần chuyển đổi thành danh sách các giao dịch (transaction).  
Hàm `dataFrame2List()` được sử dụng để loại bỏ các giá trị NaN và tạo danh sách các item trong từng giao dịch.


### 2.2. Apriori sử dụng thư viện
Sử dụng thư viện `mlxtend`:
- `apriori()` để tìm frequent itemsets
- `association_rules()` để sinh luật kết hợp

Ngưỡng sử dụng:
- min_support = 0.6


### 2.3. Apriori tự cài đặt
Thuật toán Apriori được xây dựng lại gồm các bước:

- Tính support cho từng itemset
- Sinh tập ứng viên (candidate generation)
- Loại bỏ các tập không thỏa mãn min_support
- Lặp lại cho đến khi không còn tập phổ biến

Ngoài ra, luật kết hợp được sinh bằng cách:
- Tách itemset thành antecedent và consequent
- Tính confidence
- Lọc theo min_conf
