# **1. Cấu trúc các lớp**
* **`DataLoader`**: Chịu trách nhiệm đọc dữ liệu từ file `data.csv`. 
* **`KMeansClustering`**: Lớp cơ sở thực thi thuật toán K-means tiêu chuẩn.
    * **`_initiate_centroids`**: Khởi tạo ngẫu nhiên $K$ tâm cụm từ tập dữ liệu.
    * **`_distance`**: Tính toán khoảng cách Euclidean (L2 norm).
    * **`_centroid_assignation`**: Gán từng điểm dữ liệu vào tâm cụm gần nhất dựa trên sai số tối thiểu.
    * **`fit`**: Vòng lặp chính để cập nhật vị trí tâm cụm bằng giá trị **trung bình (mean)** cho đến khi đạt ngưỡng hội tụ `tol`.
* **`KMediansClustering`**: Kế thừa từ `KMeansClustering`, thực hiện các thay đổi cốt lõi cho thuật toán K-medians:
    * **`_distance`**: Ghi đè hàm tính khoảng cách bằng khoảng cách **Manhattan (L1 norm)**.
    * **`fit`**: Ghi đè phương thức cập nhật vị trí tâm cụm bằng giá trị **trung vị (median)** dọc theo mỗi chiều không gian.

# **2. Các hàm hỗ trợ và thực thi**
* **`plot_elbow_method`**: Thực thi phương pháp "Khuỷu tay" (Elbow method) bằng cách chạy thuật toán với số lượng cụm $K$ từ 1 đến 10, giúp xác định số cụm tối ưu dựa trên tổng sai số (Total Error).
* **`Part1` & `Part2`**: Các hàm bao (wrappers) để thực thi tuần tự việc huấn luyện mô hình, in kết quả vị trí tâm cụm cuối cùng và hiển thị biểu đồ phân cụm cũng như đồ thị Elbow cho từng thuật toán tương ứng.

# **3. Điểm khác biệt chính trong triển khai**
| Đặc điểm | K-Means | K-Medians |
| :--- | :--- | :--- |
| **Hàm khoảng cách** | Euclidean ($L_2^2$) | Manhattan ($L_1$)|
| **Cập nhật tâm** | Trung bình cộng (Mean)| Trung vị (Median)|
| **Độ bền vững** | Nhạy cảm với nhiễu (outliers) | Ít bị ảnh hưởng bởi nhiễu hơn |
